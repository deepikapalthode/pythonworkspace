```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=[2,4,5,9,1]
plt.plot(x,y,'*--r',markersize=10.5,linewidth=2.5)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title("X vs Y Line Graph")
plt.show(block=False)
```

```python
import matplotlib.pyplot as plt
def plotFuncs(a,b,step):
    nSteps = int((b-a)/step)
    x = [a+ step*i for i in range(nSteps+1)]
    y1 = [t**2 for t in x]
    y2 = [t**3 for t in x]
    plt.plot(x,y1,'ro-',label="x vs x**2")
    plt.plot(x,y2,'^-g',label = "x vs x***3")
    plt.title("multiple graphs")
    plt.legend()
    plt.show()
    
def main():
    a = float(input("enter lower bound"))
    b = float(input("enter upper bound"))
    step = float(input("enter step size"))
    plotFuncs(a,b,step)
    
if __name__=="__main__":
    main()
```

```python
#subplots
import matplotlib.pyplot as plt

def plotSubplots(x):
    y1 = [i**1 for i in x]
    y2 = [i**2 for i in x]
    y3 = [i**3 for i in x]
    y4 = [i**4 for i in x]
    y5 = [i**5 for i in x]
    y6 = [i**6 for i in x]
    plt.subplot(2,3,1)  #subplot(rownum,colnum,fignum)
    plt.plot(x,y1,'ro-',label="x vs x")
    plt.title("x vs x")
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("x")
    
    plt.subplot(2,3,2)
    plt.plot(x,y2,'b<-', label ="x vs x**2")
    plt.legend()
    plt.title("x vs x**2")
    plt.xlabel("x")
    plt.ylabel("x**2")
    
    plt.subplot(2,3,3)
    plt.plot(x,y3,'g*-',label="x vs x**3")
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("x**3")
    plt.title("x vs x**3")
    
    plt.subplot(2,3,4)  #subplot(rownum,colnum,fignum)
    plt.plot(x,y4,'kx:',label="x vs x**4")
    plt.title("x vs x**4")
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("x**4")
    
    plt.subplot(2,3,5)
    plt.plot(x,y5,'mv-', label ="x vs x**5")
    plt.legend()
    plt.title("x vs x**5")
    plt.xlabel("x")
    plt.ylabel("x**5")
    
    plt.subplot(2,3,6)
    plt.plot(x,y6,'cp-.',label="x vs x**6")
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("x**6")
    plt.title("x vs x**6")
    
    plt.tight_layout()
    plt.show()
    plt.savefig("x vs x powers subplots")
    
plotSubplots(range(1,25))
    
```

```python
#histograms
import matplotlib.pyplot as plt
def plotHist(data):
    plt.hist(data,bins=[i-0.5 for i in range(0,max(data)+2)])
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title("Histogram")
    plt.xlim(min(data)-1,max(data)+1)
    plt.show()
def main():
    data = eval(input("enter data for histogram"))
    plotHist(data)
    
if __name__=="__main__":
    main()
```

```python
#piechart
import matplotlib.pyplot as plt
def plotPieChrt(data,labels):
    plt.pie(data,labels=labels,autopct='%.2f')
    plt.title("Sports Pie chart")
    plt.show()

def main():
    data = eval(input("enter sports data list"))
    labels = eval(input("enter labels"))
    plotPieChrt(data,labels)

if __name__=="__main__":
    main()
```
